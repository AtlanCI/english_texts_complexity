```python
import pandas as pd
import string
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer
from nltk import word_tokenize
from nltk.corpus import wordnet
```

```python
texts = pd.read_excel("texts.xlsx")
texts.head()
```

```python
basic_vocabulary = []
with open("A1_vocab_processed.txt", "r") as voc:
    for word in voc.readlines():
        basic_vocabulary.append(word[:-1])
basic_vocabulary = set(basic_vocabulary)
#basic_vocabulary
```

```python
text = word_tokenize("And now for received something completely different")
pos_tag(text)
```

```python
def get_wordnet_pos(treebank_tag):

    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return ''
```

```python
exclude = set(string.punctuation)
def preprocess (line, lemmatizer):
    final_split = []
    for ex_el in line.split():
        fin_elements = ex_el.split('/')
        fin_elements = [''.join(ch for ch in el if ch not in exclude) for el in fin_elements]
        #fin_elements = [el.replace('...', '') for el in fin_elements]
        fin_elements = [el.lower() for el in fin_elements]
        print(fin_elements)
        fin_elements = [lemmatizer.lemmatize(el) for el in fin_elements]
        
        final_split.extend(fin_elements)
    try:
        final_split.remove("etc.")
    except:
        pass
    return final_split
```

```python
print(lemmatizer.lemmatize("received", pos = wordnet.VERB))
```

```python
lemmatizer = WordNetLemmatizer()
processed_line = preprocess(texts["Text"][0], lemmatizer)
processed_line_set = set(processed_line)
basic_words = []
out_of_bacis_words = []
for word in processed_line:
    if word in basic_vocabulary:
        basic_words.append(word)
    else:
        out_of_bacis_words.append(word)
```
